```python
import numpy as np
import matplotlib.pyplot as plt

#exacta


A=0.01
l=0.3
k=389.6
T1_0= 400
T2_0=200
Moles=1
R=8.314
cv=(3*R)/2
C=(k*A)/(Moles*cv*l)

#condiciones iniciales

c_i=np.array([-C*(T1_0-T2_0),C*(T1_0-T2_0)])

N = 100

t = np.linspace(0.,30.,N)

#Exacta
e=np.exp(1)



def fe1(T1):
    y= (-208*(e**(-2*1.04*T1)))
    return y
def fe2(T1):
    y= (208*(e**(-2*1.04*T1)))
    return y

imagen1= fe1(t)
imagen2= fe2(t)



def f0(T1,T2):
    return -C*(T1-T2)

def f1(T1,T2):
    return C*(T1-T2)

#Euler mejorado

def GetEuler1(f0,f1,r0,t):
    
    h = (t[-1] - t[0])/(len(t)-1)
    
    T1 = np.zeros(len(t))
    T2 = np.zeros(len(t))
    
    T1[0] = c_i[0]
    T2[0] = c_i[1]
    
    K1 = np.zeros(2)
    K2 = np.zeros(2)
    
    for i in range(1,len(t)):
        
        K1[0] = f0(T1[i-1],T2[i-1])
        K1[1] = f1(T1[i-1],T2[i-1])
        
        K2[0] = f0(T1[i-1] + h*K1[0],T2[i-1] + h*K1[1])
        K2[1] = f1(T1[i-1] + h*K1[0],T2[i-1] + h*K1[1])
              
        T1[i] = T1[i-1] + 0.5*h*(K1[0]+K2[0])
        T2[i] = T2[i-1] + 0.5*h*(K1[1]+K2[1]) 
        
    return T1,T2

T1,T2 = GetEuler1(f0,f1,c_i,t)


fig = plt.figure(figsize=(10,10))
plt.title("Soluci贸n con "+str(N)+" pasos")
ax = fig.add_subplot(111)
ax.plot(t,T1, label='T1(t) Euler mejorado')
ax.plot(t,T2, label='T2(t) Euler mejorado')
ax.plot(t,imagen1, label='Soluci贸n exacta T1')
ax.plot(t,imagen2, label='Soluci贸n exacta T2')
ax.legend()
plt.savefig("Soluci贸necuacionesJoanFaber.png")
```
