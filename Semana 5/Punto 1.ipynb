```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=5000
t=np.linspace(0,10,N)
```

```python
##definimos las funciones
```

```python
def exacta(t):
    fnc=np.exp(t)
    return fnc

def f(u,q):
   fn=u**q
   return fn
```

```python
##solucionamos con rungekutta 4
```

```python
def Rungekutta(q,r0):

    h = (t[-1] - t[0])/(len(t)-1)
    
    u=np.zeros(len(t))
    u[0]=r0
    
    K1 = np.zeros(4)
    K2 = np.zeros(4)
    K3 = np.zeros(4)
    K4 = np.zeros(4)
    for i in range(1,len(t)):
        K1 = f(u[i-1],q)
        K2 = f(u[i-1]+(h/2)*K1,q)
        K3 = f(u[i-1]+(h/2)*K2,q)
        K4 = f(u[i-1]+(h)*K3,q)
        u[i]= u[i-1] + (h/6)*(K1+2*K2+2*K3+K4)
        
        
    return u
```

```python
q=np.array([0,0.2,0.4,0.7,0.9,1])
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)

for i in range(len(q)):
    y2= Rungekutta(q[i],1)
    ax.plot(t,y2,label = str(q[i]))
    ax.legend()  
```
